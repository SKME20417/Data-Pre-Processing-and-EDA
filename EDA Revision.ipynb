```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
pd.set_option('display.max_columns',None)
```

```python
df  = pd.read_csv('train.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
feature_with_nan= [feature for feature in df.columns if df[feature].isnull().sum()>=1]
```

```python
feature_with_nan
```

```python
#df['Electrical'].isnull().mean()
```

```python
for features in feature_with_nan:
    print(features,np.round(df[features].isnull().mean()*100,4),'% missing values')
```

```python
import warnings
warnings.filterwarnings('ignore')

sns.boxplot('SalePrice', data=df)
plt.show()
```

```python
data = df.copy()

for i in feature_with_nan:
    data[i] = np.where(data[i].isnull(),1,0)
    data.groupby(i)['SalePrice'].median().plot.bar()
    plt.title(i)
    plt.show()
```

```python
numerical_feature = [i for i in df.columns if df[i].dtypes!='O']
```

```python
print(len(numerical_feature))
```

```python
df[numerical_feature].head()
```

```python
df.info()
```

```python
year_feature = [i for i in numerical_feature if "Yr" in i or 'Year' in i]
```

```python
for i in year_feature:
    print(i,df[i].unique())
```

```python
df.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('year sold')
plt.ylabel('meadian house price')
plt.title('house price v/s year sold')
plt.show()
```

```python
data  = df.copy()

for i in year_feature:
    if i != 'YrSold':
        data[i] = data['YrSold']-data[i]
        plt.scatter(data[i],data['SalePrice'])
        plt.xlabel(i)
        plt.ylabel('sales price')
        plt.show()
```

```python
discreate_feature = [i for i in numerical_feature if len(df[i].unique())<=25]
```

```python
print(len(discreate_feature))
```

```python
df[discreate_feature].head()
```

```python
data = df.copy()
for i in discreate_feature:
    data.groupby(i)['SalePrice'].median().plot.bar()
    plt.xlabel(i)
    plt.ylabel('sales price')
    plt.title(i)
    plt.show()
    
```

```python
continuous_feature = [i for i in numerical_feature if i not in discreate_feature+year_feature] 
```

```python
print(len(continuous_feature))
```

```python
data = df.copy()

for i in continuous_feature:
    data.groupby(i)['SalePrice'].median().plot.hist()
    plt.title(i)
    plt.xlabel(i)
    plt.ylabel('Count')
    plt.grid()
    plt.show()
```

```python
data = df.copy()
for i in continuous_feature:
    if 0 in data[i].unique():
        pass
    else:
        data[i]= np.log(data[i])
        plt.scatter(data[i], data['SalePrice'])
        plt.xlabel(i)
        plt.ylabel('sales price')
        plt.title(i)
        plt.show()
```

```python
data = df.copy()

for i in continuous_feature:
    if 0 in data[i].unique():
        pass
    else:
        data[i] = np.log(data[i])
        data.boxplot(column=i)
        plt.xlabel(i)
        plt.ylabel(i)
        plt.show()
```
